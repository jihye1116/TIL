```python
#시각화 모듈
import matplotlib.pyplot as plt
```

```python
#부산 3월 기온
temp = [10.8, 8.5, 8.7, 7.9, 5.1, 
        8.1, 8.2, 10.4, 11, 9.9, 
        7.8, 9.5, 11.6, 8.2, 7.7, 
        6.6, 11.1, 12.2, 12.6, 10.4, 
        13.7, 14.1, 12.6, 11.5, 13, 
        14.5, 13.7, 9.4, 10.4, 11.6, 12.9]
```

```python
# STEP 2. 코랩(Colab)의 런타임을 재시작
# STEP 3. matplotlib의 폰트를 Nanum 폰트로 지정
plt.rc('font', family="NanumBarunGothic")
```

## **선 옵션**
+ **색상**
  - b, g, r, c, y, l, w
+ **마커**
  - o(원), v(역삼각형), s(네모), +(플러스), .(점)
+ **선스타일*
  - \-(soild line), --(dashed line), -.(dash-dot), :(dotted line)


```python
# 그래프 그리기. 제목 달기
plt.title('3월 기온')
plt.plot(temp)
plt.show()
```

```python
# 그래프 크기 조절: 가로 세로를 튜플로
plt.figure(figsize=(10,6))

plt.title('3월 기온')
plt.xlabel('날짜')
plt.ylabel('기온')

plt.grid(True, color='gray', alpha=0.5, linestyle='--') # 그리드
plt.axhline(y=10, color='r', linewidth=1)
plt.axvline(x=5, color='r', linewidth=1)

plt.plot(temp, 'go--') #선옵션
plt.show()
```

```python
# 그래프 크기 조절: 가로 세로를 튜플로
plt.figure(figsize=(10,6))

plt.title('3월 기온')
plt.xlabel('날짜')
plt.ylabel('기온')

plt.grid(True, color='gray', alpha=0.5, linestyle='--') # 그리드
plt.axhline(y=10, color='r', linewidth=1)
plt.axvline(x=5, color='r', linewidth=1)

plt.plot(temp, 'g--', label= '기본') #그래프1
plt.plot(temp, 'b--', drawstyle='steps-post',label= '계단식') #그래프2

plt.show()
```

```python
# 여러 그래프 그리기
# 그래프 그리기
x = list(range(1,32))

plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.plot(x, temp, 'go--')
plt.title('기본 그래프')

#행 열 첫
plt.subplot(2,2,2)
plt.scatter(x, temp)
plt.title('산포도 그래프')

plt.subplot(2,2,3)
plt.bar(x, temp)
plt.title('막대 그래프')

plt.subplot(2,2,4)
plt.barh(x, temp)
plt.title('가로막대 그래프')

plt.show()
```
